```python
# feature target separation x and y

x = df.drop('attrition',axis=1)
y = df['attrition']
```

```python
x.columns,y.name
```

```python
# identifying column types

cat_cols = x.select_dtypes(include = 'object').columns
num_cols = x.select_dtypes(exclude = 'object').columns
```

```python
# importing sklearn modules columntransformer , onehotencoder
# and standardscaler

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',drop='first')

preprocessor = ColumnTransformer(
    transformers =[
        ('cat',ohe,cat_cols),
        ('num',StandardScaler(),num_cols)
    ]
)
```

```python
# train_test_split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42,stratify=y
)
```

```python
# creating pipelines with model
# so preprocessing + model training

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
pipe_lr = Pipeline([
    ('preprocessor',preprocessor),
    ('model',LogisticRegression())
])

pipe_lr.fit(x_train,y_train)
```

```python
# decision tree

from sklearn.tree import DecisionTreeClassifier

pipe_dt = Pipeline([
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(random_state=42))
])

pipe_dt.fit(x_train,y_train)
```

```python
# random forest pipeline

from sklearn.ensemble import RandomForestClassifier

pipe_rf = Pipeline([
    ('preprocessor',preprocessor),
    ('model',RandomForestClassifier(random_state=42,n_estimators=200))
])

pipe_rf.fit(x_train,y_train)
```

```python
# linear regression
y_pred_lr_train = pipe_lr.predict(x_train)
y_pred_lr_test = pipe_lr.predict(x_test)

# decision tree
y_pred_dt_train = pipe_dt.predict(x_train)
y_pred_dt_test = pipe_dt.predict(x_test)

# random forest
y_pred_rf_train = pipe_rf.predict(x_train)
y_pred_rf_test = pipe_rf.predict(x_test)

```

```python
from sklearn.metrics import accuracy_score,
confusion_matrix,classification_report
```

```python
# linear regression accuracy
train_acc_lr = accuracy_score(y_train,y_pred_lr_train)
test_acc_lr = accuracy_score(y_test,y_pred_lr_test)
print(train_acc_lr)
print(test_acc_lr)
```

```python
# decision tree accuracy
train_acc_dt = accuracy_score(y_train,y_pred_dt_train)
test_acc_dt = accuracy_score(y_test,y_pred_dt_test)
print(train_acc_dt)
print(test_acc_dt)
```

```python
# random forest
train_acc_rf = accuracy_score(y_train,y_pred_rf_train)
test_acc_rf = accuracy_score(y_test,y_pred_rf_test)
print(train_acc_rf)
print(test_acc_rf)
```

```python
# linear , decsion , random

train_acc = [train_acc_lr,train_acc_dt,train_acc_rf]
test_acc = [test_acc_lr,test_acc_dt,test_acc_rf]
print(train_acc)
print(test_acc)
```

```python
print('linear regression')
print(classification_report(y_test,y_pred_lr_test))
```

```python
print('decision tree')
print(classification_report(y_test,y_pred_dt_test))
```

```python
print('random forest')
print(classification_report(y_test,y_pred_rf_test))
```

```python
print('logistic regression')
print(confusion_matrix(y_test,y_pred_lr_test))

```

```python
print('decision tree')
print(confusion_matrix(y_test,y_pred_dt_test))
```

```python
print('random forest')
print(confusion_matrix(y_test,y_pred_rf_test))
```

```python
from sklearn.metrics import recall_score

print("logistic regression recall:", recall_score(y_test,y_pred_lr_test))
print("decision tree recall:", recall_score(y_test,y_pred_dt_test))
print("random forest recall:", recall_score(y_test,y_pred_rf_test))
#
```

```python
from sklearn.metrics import accuracy_score , f1_score

models={
    'logistic regression':y_pred_lr_test,
    'decision tree':y_pred_dt_test,
    'random forest':y_pred_rf_test
}

for name,pred in models.items():
  print(name)
  print('accuracy:',accuracy_score(y_test,pred))
  print('f1 score:',f1_score(y_test,pred))
  print('recall:',recall_score(y_test,pred))
  print('---------------------------------')
```
